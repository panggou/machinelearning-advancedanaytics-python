#### Decision Tree
Data: Iris dataset

```python
#Import the libraries

import pandas as pd
from sklearn import datasets, metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#load the data

iris = datasets.load_iris()

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

iris_df.head()
```

```python
#X and y

X = iris_df.iloc[:,:].values
#or X = iris.data

y= iris.target
```

```python
#Splitting the data into training and testing

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#import the tree library

from sklearn import tree
```

```python
#create an instance of the decision tree

dtree = tree.DecisionTreeClassifier()
```

```python
#train the model
dtree.fit(X_train, y_train)
```

```python
#plot the descision tree

plt.figure(figsize = (25,10)) #big tree to be plot

dtree_plot = tree.plot_tree(dtree, feature_names=iris.feature_names,
                           class_names=iris.target_names,
                           filled = True,
                           fontsize=14)
plt.show()

```

```python
#prediction of the test data

y_pred = dtree.predict(X_test)
```

```python
#accuracy score

acc = metrics.accuracy_score(y_test, y_pred)

acc
```

```python
#if there is a new flower with the following specifications

my_flower = [1.1, 0.5, 9.4, 7.2]

pred_flower = dtree.predict([my_flower])

my_flower_list = list(iris.target_names)

pred_flower
```
